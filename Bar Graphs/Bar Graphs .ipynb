```python
from matplotlib import pyplot as plt
x = ['inception V3','resNet 50' , 'resNet 152V2']
y = [98.33,71.67,100]
fig = plt.figure(figsize = (5, 7))
barplot = plt.bar(x, y)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.5, yval, (yval), va='bottom') #va: vertical alignment y positional argument
    
plt.title("Training Accuracy(20 Epochs)")
plt.xlabel('Models')
plt.ylabel("Accuracy(%)")

plt.savefig('ModelsTrainingAccuracy.jpg')
```

```python
from matplotlib import pyplot as plt
models = ['Inception V3','ResNet 50', 'ResNet 152V2' ]
train_acc = [98.33,71.67,100.00]
valid_acc = [90.00,66.66,96.67]
#fig = plt.figure(figsize = (5, 7))
plt.plot(range(0,1))
plt.plot(models, train_acc,label='training accuracy' ,color='g')
plt.plot(models, valid_acc,label='validation accuracy' ,color='orange')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy Of Models For 20 Epochs')
plt.legend()
plt.show()
plt.savefig('ModelsLineGraphAccVsVal.jpg')
```

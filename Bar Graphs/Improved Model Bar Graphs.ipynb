```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('Original inception V3', 'Improved inception V3')
y_pos = np.arange(len(objects))
performance = [94.22,98.98]
fig = plt.figure(figsize = (4.5, 10))
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Training Accuracy')

plt.show()
```

```python
from matplotlib import pyplot as plt
```

```python
x = ['Original inception V3', 'Improved inception V3']
y = [94.22,98.98]
```

```python
fig = plt.figure(figsize = (5, 7))
barplot = plt.bar(x, y)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.5, yval, (yval), va='bottom') #va: vertical alignment y positional argument
    
plt.title("Training Accuracy")
plt.xlabel('Models')
plt.ylabel("Accuracy(%)")

plt.savefig('InceptionTrainingAccuracy.jpg')
```
